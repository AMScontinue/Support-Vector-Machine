```python
from matplotlib import pyplot as plt
import numpy as np
import random
import time
import csv
def sgd_svm(x,y,opts): 
    start_time = time.time() 
    max_count = opts[0] 
    C = opts[1] 
    D = x.shape[0] 
    N = x.shape[1] 
    w = np.zeros(D) 
    b = 0;
    w = np.append(w,b) 
    x = np.append(x,[np.ones(N)],axis=0) 
    sample_num = int(N) 
    sample_list = [s for s in range(N)] 
    for j in range(1,max_count): #开始迭代
        sample_list = random.sample(sample_list,sample_num) 
        sample_x = x[:,sample_list] 
        sample_y = y[sample_list] 

        kexi = np.maximum(np.zeros(sample_num),np.ones(sample_num)-sample_y*(w@sample_x))

        g = w[0:D]*2/(sample_num*C)-(sample_y[np.nonzero(kexi)[0]]@sample_x[0:D,np.nonzero(kexi)[0]].T)/sample_num

        g1 = -(sample_y[np.nonzero(kexi)[0]]@sample_x[D,np.nonzero(kexi)[0]].T)/sample_num
        g = np.append(g,g1) 
        w = w-g 
    end_time = time.time() 
    print('Took %f second' % (end_time - start_time)) 
    return w #返回 w


xt=[]
yt=[]
with open("/kaggle/input/lab1temp/X_train.csv", mode="r", encoding="utf-8-sig") as f:
    
    reader = csv.reader(f)
    header = next(reader)

    for row in reader:
        xt.append([row[0],row[1],row[2],row[3],row[4],row[5],row[6],row[7],row[8],row[9],row[10],row[11],row[12],row[13],row[14],row[15],row[16],row[17],row[18],row[19],row[20],row[21],row[22],row[23],row[24],row[25],row[26],row[27],row[28]])

with open("/kaggle/input/lab1temp/Y_train.csv", mode="r", encoding="utf-8-sig") as g:
    reader = csv.reader(g)
    header = next(reader)

    for row in reader:
        yt.append(row[0])

ytt=[]
i = 0 
while i < len(yt) :
    if yt[i] == '0':
        ytt.append(-1)
    else:
        ytt.append(1)
    i=i+1
    
    
x = np.array(xt,dtype = float)
y = np.array(ytt,dtype = float)
x=x.T

opts=(10000,100) #优化参数，第一项表示迭代次数，第二项表示 C 的值
w = sgd_svm(x,y,opts) 
print("final w:",w) #打印最优解

#上面是svm求解，下面是测试

xtestt=[]
ytestt=[]
with open("/kaggle/input/lab1temp/X_test.csv", mode="r", encoding="utf-8-sig") as f:
    
    reader = csv.reader(f)
    header = next(reader)

    for row in reader:
        xtestt.append([row[0],row[1],row[2],row[3],row[4],row[5],row[6],row[7],row[8],row[9],row[10],row[11],row[12],row[13],row[14],row[15],row[16],row[17],row[18],row[19],row[20],row[21],row[22],row[23],row[24],row[25],row[26],row[27],row[28]])

with open("/kaggle/input/lab1temp/Y_test.csv", mode="r", encoding="utf-8-sig") as g:
    reader = csv.reader(g)
    header = next(reader)

    for row in reader:
        ytestt.append(row[0])

ytesttt=[]        
i = 0 
while i < len(ytestt) :
    if ytestt[i] == '0':
        ytesttt.append(-1)
    else:
        ytesttt.append(1)
    i=i+1

testx = np.array(xtestt,dtype = float)
testy = np.array(ytesttt,dtype = float)
temp=[np.ones(testx.shape[0])]
testx=testx.T
testx = np.append(testx,temp,axis=0)

result = np.matmul(w,testx)
#print(result)

fact=result*testy
positive=0
for f in fact:
    if f>0:
        positive=positive+1
        
print("positive:",positive)
print("positive rate:",positive/len(ytestt))



        
```
